```python
# Human
import pandas as pd

tissues = ['heart', 'lung', 'stomach']

for t in tissues:
	svm_results = pd.read_csv(f"human/SVM_predictions_{t}.csv")
	mlp_results = pd.read_csv(f"human/MLP_predictions_{t}.csv")
	
	svm_ess = svm_results[svm_results['Pre_Label']==1]
	mlp_ess = mlp_results[mlp_results['Pre_Label']==1]

	print(mlp_ess.shape[0])

	ess = pd.merge(svm_ess, mlp_ess, on='lncRNA_id', how='inner')
	ess = ess[['lncRNA_id']]
	ess.to_csv(f"./human/{t}_essential_genes.csv", index=False)
```

```python
# Mouse
import pandas as pd

tissues = ['heart', 'lung', 'brain']

for t in tissues:
	svm_results = pd.read_csv(f"mouse/SVM_predictions_{t}.csv")
	mlp_results = pd.read_csv(f"mouse/MLP_predictions_{t}.csv")
	
	svm_ess = svm_results[svm_results['Pre_Label']==1]
	mlp_ess = mlp_results[mlp_results['Pre_Label']==1]
	print(mlp_ess.shape[0])

	ess = pd.merge(svm_ess, mlp_ess, on='lncRNA_id', how='inner')
	ess = ess[['lncRNA_id']]
	ess.to_csv(f"./mouse/{t}_essential_genes.csv", index=False)
```
